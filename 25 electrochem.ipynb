```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
```

```python
conc = np.array([0.0315, 0.01562, 0.00781, 0.00391, 0.00195, 0.00098])
eq_conduc = np.array([9.19, 12.94, 18.17, 25.43, 35.5, 49.12])
```

```python
# Configure plot settings
plt.rc({
    'figure.figsize': (8, 6),
    'font.family': 'Times New Roman',
    'font.size': 25
})

# Optimized interpolation function
def interpol(x,y, degree, mark, color, colorline):
    coeffs = np.polyfit(x, y, degree)
    x_fit = np.linspace(0,x[0],100)  
    y_fit = np.polyval(coeffs, x_fit)
    plt.plot(x_fit, y_fit, c=colorline , linestyle = '-')
    return coeffs

# Perform polynomial fits
initial_fit = interpol(conc*eq_conduc, 1/eq_conduc, 1, 's', 'r','b')

#Plot parameters
plt.title(r'Графический метод определения $\Lambda_{0}$', fontsize=10 )
plt.xlabel(r'$\Lambda$*C')
plt.ylabel(r'1/$\Lambda_{0}$')
plt.scatter(conc*eq_conduc, 1/eq_conduc, marker = 'D', c='r',  )
plt.grid(linestyle = '--')
plt.xlim(0, 0.3 )

print(initial_fit)
```
