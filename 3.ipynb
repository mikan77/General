```python
import numpy as np

# Raw data lines (excluding headers)
raw_data = [
    "0 0 80,7",
    "6,0 16,4 77,0",
    "13,1 25,3 74,7",
    "17,5 29,9 73,7",
    "24,7 35,6 72,6",
    "36,9 42,0 71,8",
    "48,0 46,5 71,5",
    "64,5 57,4 72,1",
    "73,6 63,4 73,0",
    "80,4 69,8 74,0",
    "87,6 77,7 75,3",
    "100 100 79,6"
]

# Initialize lists to hold each column's data
x1 = []  # Liquid phase mole %
y1 = []  # Vapor phase mole %
temps = []  # Boiling temperatures

# Process each line to extract values
for line in raw_data:
    parts = line.split()
    # Convert each value by replacing commas with periods and converting to float
    x1.append(float(parts[0].replace(',', '.')))
    y1.append(float(parts[1].replace(',', '.')))
    temps.append(float(parts[2].replace(',', '.')))

# Create NumPy arrays
x1_array = np.array(x1)
y1_array = np.array(y1)
temp_array = np.array(temps)

# Resulting arrays:
print("x1 (Liquid phase):", x1_array)
print("y1 (Vapor phase):", y1_array)
print("Temperatures (°C):", temp_array)
```

```python
import matplotlib 
import matplotlib.pyplot as plt
```

```python
plt.rcParams["font.family"] = "Times New Roman"
matplotlib.rcParams.update({
            'font.size': 16,
            'figure.figsize': (8, 6)
        })


inte1 = np.polyfit(x1_array, temp_array, 6)
inte2 = np.polyfit(y1_array, temp_array, 6)
plt.scatter(x1_array, temp_array)
plt.scatter(y1_array, temp_array)
x_fit = np.linspace(0,100,100)
y_fit_1 = np.polyval(inte1, x_fit)
y_fit_2 = np.polyval(inte2, x_fit)
plt.plot(x_fit, y_fit_1, label='Calibration Line ')
plt.plot(x_fit, y_fit_2, label='Calibration Line ')
plt.axhline(y=74, color='r', linestyle='-')
plt.ylim(70,)
plt.xlim(0,100)
plt.xticks(np.arange(0, 100, step=10))
plt.xlabel('Мольный % метилэтилкетона', fontsize=14)
plt.ylabel('Температура, C', fontsize=14)
plt.grid()
plt.show
```
