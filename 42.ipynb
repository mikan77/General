```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python


# Ваши данные (пример DataFrame)
data = {
    'Длина волны λ, нм': [400, 425, 450, 475, 500, 525, 550, 575, 600, 700, 750],
    '0.0001': [0.60, 0.67, 0.72, 0.71, 0.66, 0.56, 0.40, 0.31, 0.25, 0.08, 0.05],
    '0.000075': [0.45, 0.51, 0.54, 0.51, 0.48, 0.42, 0.30, 0.24, 0.18, 0.06, 0.03],
    '0.00005': [0.30, 0.33, 0.36, 0.35, 0.33, 0.28, 0.20, 0.16, 0.12, 0.04, 0.02],
    '0.000025': [0.15, 0.17, 0.18, 0.17, 0.16, 0.14, 0.10, 0.08, 0.06, 0.02, 0.01]
}

df = pd.DataFrame(data)

# 1. Извлекаем данные для 750 нм
row_750nm = df[df['Длина волны λ, нм'] == 750].iloc[0]

# 2. Подготавливаем данные для графика
concentrations = [0.0001, 0.000075, 0.00005, 0.000025]  # моль/л
absorbances = [row_750nm['0.0001'], 
               row_750nm['0.000075'], 
               row_750nm['0.00005'], 
               row_750nm['0.000025']]

# 3. Строим график
plt.figure(figsize=(8, 5))
plt.plot(concentrations, absorbances, 'bo-', label='750 нм')
plt.xlabel('Концентрация, моль/л', fontsize=12)
plt.ylabel('Оптическая плотность', fontsize=12)
plt.title('Зависимость оптической плотности от концентрации (λ = 750 нм)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
epsilon_10 = []
for v in data['0.0001']:
    epsilon_10.append(v/0.0001)

epsilon_7_5 = []
for v in data['0.000075']:
    epsilon_7_5.append(v/0.0001)

epsilon_5 = []
for v in data['0.00005']:
    epsilon_5.append(v/0.0001)

epsilon_2_5 = []
for v in data['0.000025']:
    epsilon_2_5.append(v/0.0001)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(df['Длина волны λ, нм'],epsilon_10 , 'bo-', label=r'$10\cdot10^{-5}$ M')
plt.plot(df['Длина волны λ, нм'],epsilon_7_5 , 'bo-', label=r'$7.5\cdot10^{-5}$ M')
plt.plot(df['Длина волны λ, нм'],epsilon_5 , 'bo-', label=r'$5\cdot10^{-5}$ M')
plt.plot(df['Длина волны λ, нм'],epsilon_2_5 , 'bo-', label=r'$2.5\cdot10^{-5}$ M')

plt.xlabel(r'$\lambda$', fontsize=12)
plt.ylabel(r'$\epsilon$ ', fontsize=12)
#plt.title(r'Концентрация $10\cdot10^{-5}$ M')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```
