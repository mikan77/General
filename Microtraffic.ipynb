```python
from shapely.ops import transform

# Load your data
df = gpd.read_file(r"C:\Users\Brian\Desktop\py\Microtraffic\microtrafico\A_Microtraffic_Quito_Parish_2018_2021.shp")

# Define a function to swap x and y coordinates
swap_coords = lambda x, y: (y, x)  # Swap x â†” y

# Apply the coordinate swap to all geometries
df_swapped = df.copy()
df_swapped.geometry = df.geometry.apply(lambda geom: transform(swap_coords, geom))

plt.rc({
    'figure.figsize': (15, 10),
    'font.family': 'Times New Roman',
    'font.size': 25
})
# Explicitly create a figure and axes with your desired size


fig, ax = plt.subplots(figsize=(13, 10))
ax.invert_yaxis()

ax.set_ylabel('X (Rotated)')
df_swapped.plot(
    column='April2018',
    legend=True,
    ax=ax,
    legend_kwds={
        'shrink': 0.6,     # Scale the height (0.6 = 60% of the plot height)
        'aspect': 20,       # Ratio of long to short dimensions (height/width)
        'pad': 0.1,         # Padding between plot and colorbar (as fraction of plot width)
        'label': "April 2018 Values"  # Custom label
    }
)

ax.set_title('April 2018 (Swapped Coordinates)')
plt.xlabel('Northing (y), m')
plt.ylabel('Easting (x), m')
plt.show()
```
