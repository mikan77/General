```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
xa = np.array([5, 4,3,2,1, 0])
xb = np.array([0, 1,2,3,4,5 ])
conc = xa/(xa+xb)
D = np.array([1.4425, 1.4455, 1.448, 1.451, 1.454, 1.459])
```

```python
plt.rcParams["font.family"] = "Times New Roman"
matplotlib.rcParams.update({
            'font.size': 16,
            'figure.figsize': (8, 6)
        })


inte0 = np.polyfit(conc, D, 1)
plt.scatter(conc, D)
x_fit = np.linspace(0,1,100)
y_fit_0 = np.polyval(inte0, x_fit)
plt.plot(x_fit, y_fit_0, label='Calibration Line ')

plt.xlim(0,1)
#plt.xticks(np.arange(0, 5, step=1))
plt.title('График зависимости показателя преломления от объемной доли ')
plt.xlabel(r'V$_{CCl_4}$/V$_{CHCl_3}$', fontsize=14)
plt.ylabel('Показатель преломления', fontsize=14)
plt.grid()
plt.show
```
