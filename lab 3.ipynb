```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
xa = np.array([5, 4,3,2,1, 0])
xb = np.array([0, 1,2,3,4,5 ])
conc = xa/(xa+xb)
D = np.array([1.443, 1.4455, 1.448, 1.451, 1.454, 1.459])
```

```python
plt.rcParams["font.family"] = "Times New Roman"
matplotlib.rcParams.update({
            'font.size': 16,
            'figure.figsize': (8, 6)
        })


inte0 = np.polyfit(conc, D, 1)
plt.scatter(conc, D)
x_fit = np.linspace(0,1,100)
y_fit_0 = np.polyval(inte0, x_fit)
plt.plot(x_fit, y_fit_0, label='Calibration Line ')

plt.xlim(0,1)
#plt.xticks(np.arange(0, 5, step=1))
plt.xlabel(r'V/V % $CHCl_3$', fontsize=14)
plt.ylabel(', C', fontsize=14)
plt.grid()
plt.show
```

```python
exp_3 = {'temp_ini':np.array([61.2, 62, 65, 68, 72, 76.7 ]),
        'D_exp': np.array([1.44275,1.446,1.4485,1.4535]),
        'Conc': []}

for i in exp_3['D_exp']:
    conc_gas = (i - inte0[1])/inte0[0]
    exp_3['Conc'].append(conc_gas)
    print(conc_gas)
    
```

```python
# Create NumPy arrays
x1_array = conc

y1_array = np.concatenate(([1], np.array(exp_3['Conc']), [0]))
temp_array = exp_3['temp_ini']

plt.rcParams["font.family"] = "Times New Roman"
matplotlib.rcParams.update({
            'font.size': 16,
            'figure.figsize': (8, 6)
        })


inte1 = np.polyfit(x1_array, temp_array, 2)
inte2 = np.polyfit(y1_array, temp_array, 2)
plt.scatter(x1_array, temp_array)
plt.scatter(y1_array, temp_array)
x_fit = np.linspace(0,1,100)
y_fit_1 = np.polyval(inte1, x_fit)
y_fit_2 = np.polyval(inte2, x_fit)
plt.plot(x_fit, y_fit_1, label='Calibration Line ')
plt.plot(x_fit, y_fit_2, label='Calibration Line ')
#plt.axhline(y=74, color='r', linestyle='-')
plt.xlim(0,1)
plt.ylim(40,)

plt.xlabel('V/V % $CHCl_3$', fontsize=14)
plt.ylabel(', C', fontsize=14)
plt.grid()
plt.show
```
